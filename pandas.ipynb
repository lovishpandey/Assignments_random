```python
#Join users to transactions, keeping only rows from transactions and users that match via
#UserID (inner join)
pd.merge(users,transactions,on='UserID',how='inner')
```

```python
#Join users to transactions, displaying all matching rows AND all non-matching rows (full
#outer join)
pd.merge(users,transactions,on='UserID',how='outer')
```

```python
##Build a dataset with every possible (UserID, ProductID) pair (cross join)
user=pd.DataFrame({"UserID":users["UserID"]})
product=pd.DataFrame({"ProductID":products["ProductID"]})
user["key"]=1
product["key"]=1
```

```python
final=pd.merge(user,product,on='key',how='left')
final.drop(['key'],axis=1)
```

```python
##Determine how much quantity of each product was purchased by each user
users.head()
```

```python
transactions.head()
```

```python
quantity=pd.merge(final,transactions,on=['UserID','ProductID'],how='left')
quantity
```

```python
quantity.groupby(['UserID','ProductID'])['Quantity'].sum()
```

```python
data=pd.merge(users,transactions,on='UserID',how='left').drop_duplicates(subset='UserID')
data
```

```python
##Test to see if we can drop columns
my_columns = list(data.columns)
my_columns
```

```python
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns) #set threshold to drop NAs
```

```python
missing_info = list(data.columns[data.isnull().any()])
missing_info
```

```python
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing)) #count of missing data
```

```python
for col in missing_info:
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0]
    print('percent missing for column {}: {}'.format(col, percent_missing))
```
